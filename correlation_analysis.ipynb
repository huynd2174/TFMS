```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dữ liệu mới từ bảng người dùng cung cấp
data = {
    'Company': ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'TSLA'],
    'RFR': [150.89, 97.64, 94.83, 262.78, 196.80],
    'LSTM': [152.20, 100.84, 99.98, 266.33, 200.68],
    'SA': [0.081, 0.093, 0.163, 0.091, 0.135],
    'Actual Close': [153.85, 99.54, 95.00, 271.32, 194.64]
}

df = pd.DataFrame(data)

# Biểu đồ tương quan giữa RFR và Actual Close
sns.scatterplot(x='RFR', y='Actual Close', data=df)
plt.title('Correlation between RFR and Actual Close')
plt.show()

# Biểu đồ tương quan giữa LSTM và Actual Close
sns.scatterplot(x='LSTM', y='Actual Close', data=df)
plt.title('Correlation between LSTM and Actual Close')
plt.show()

# Biểu đồ tương quan giữa SA và Actual Close
sns.scatterplot(x='SA', y='Actual Close', data=df)
plt.title('Correlation between SA and Actual Close')
plt.show()

# Tính hệ số tương quan
rfr_correlation = df['RFR'].corr(df['Actual Close'])
lstm_correlation = df['LSTM'].corr(df['Actual Close'])
sa_correlation = df['SA'].corr(df['Actual Close'])

# Hiển thị kết quả
print("The correlation between RFR and Actual Close is:", rfr_correlation)
print("The correlation between LSTM and Actual Close is:", lstm_correlation)
print("The correlation between SA and Actual Close is:", sa_correlation)
```
